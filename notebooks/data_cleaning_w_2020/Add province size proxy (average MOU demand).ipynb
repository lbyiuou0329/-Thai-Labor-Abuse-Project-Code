```python
import os
import re

import pandas as pd
import numpy as np

from datetime import datetime as dt

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data_dir = '/Users/boyuliu/pyprojects/Joann/Joann-Thailand-Project/notebooks/datasets/new_dataset/'

wv_data_file = 'regression_data_wv_cases1_causal_ma_detrend_20210211.csv'
# wv_data_file_lg_prov = 'regression_data_wv_cases1_causal_detrend_province_20_20210209.csv'
print(wv_data_file)
df = pd.read_csv(data_dir + wv_data_file)
print(df.shape)
df.head()
```

```python
province_sizes = df.groupby('province').mean()['total_demand'].reset_index().rename(
    columns={'total_demand':'province_size'})
province_sizes
```

```python
df = df.merge(province_sizes, on='province')
df
```

```python
df.to_csv(data_dir + wv_data_file, index=False)
```
